```typescript
import { HfInference } from "npm:@huggingface/inference";
```

```typescript
const HF_TOKEN = "hf_..."
```

You can also omit "model" to use the recommended model for the task

```typescript
const inference = new HfInference(HF_TOKEN);

await inference.translation({
  model: 't5-base',
  inputs: 'What am I going to do after this'
})
```

```typescript
import { display } from "https://deno.land/x/display/mod.ts";
```

```typescript
const image = await inference.textToImage({
  model: 'stabilityai/stable-diffusion-2',
  inputs: 'award winning high resolution photo of a giant bird flying in the sky, [trending on artstation]',
  parameters: {
    negative_prompt: 'blurry',
  }
})

const b = (blob, callback) => {
    var a = new FileReader();
    a.onload = function(e) {callback(e.target.result);}
    a.readAsDataURL(blob);
}
b(image, async (result) => {
    await display({"image/jpeg": result.split(',')[1]}, {raw: true})
})
```

```typescript
await inference.imageToText({
  data: image,
  model: 'nlpconnect/vit-gpt2-image-captioning',  
})
```
